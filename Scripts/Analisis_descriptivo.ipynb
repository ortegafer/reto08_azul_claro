```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


paleta = {
    'A': '#e2211c',  
    'B': '#FF6F61',
    'C': '#FF4500',
    'D': '#8B0000' 
}

compresores = ['A', 'B', 'C', 'D']

stats = df.groupby('Compresor').describe().transpose()
stats
```

```python
plt.figure(figsize=(10, 6))

for i, comp in enumerate(compresores):
    subset = df[df['Compresor'] == comp]['Presion']
    plt.boxplot(subset, positions=[i], patch_artist=True,
               boxprops=dict(facecolor=paleta[comp], alpha=0.7),
               medianprops=dict(color='black'),
               whiskerprops=dict(color=paleta[comp]),
               capprops=dict(color=paleta[comp]),
               flierprops=dict(markeredgecolor=paleta[comp]))

plt.title('Distribución de Presión por Compresor', fontsize=14)
plt.xlabel('Compresor')
plt.ylabel('Presión (bar)')
plt.xticks(range(len(compresores)), compresores)
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
plt.figure(figsize=(16, 12))

for i, comp in enumerate(compresores, 1):
    plt.subplot(2, 2, i)
    subset = df[df['Compresor'] == comp]
    
    sns.kdeplot(subset['Potencia_Medida'], color=paleta[comp], 
                linestyle='-', linewidth=2.5, label='Medida', alpha=0.8)
    
    sns.kdeplot(subset['Potencia_Estimada'], color=paleta[comp], 
                linestyle='--', linewidth=2.5, label='Estimada', alpha=0.8)
    
    plt.fill_betweenx(np.linspace(0, 0.1, 100), 
                      subset['Potencia_Medida'].min(), 
                      subset['Potencia_Medida'].max(),
                      color=paleta[comp], alpha=0.1)
    
    plt.title(f'Compresor {comp}', fontsize=13, pad=10, color=paleta[comp])
    plt.xlabel('Potencia (kW)', fontsize=11)
    plt.ylabel('Densidad', fontsize=11)
    plt.legend(framealpha=1, edgecolor='#ddd')
    plt.grid(True, linestyle='--', alpha=0.3)
    
    stats_text = (f"Medida μ: {subset['Potencia_Medida'].mean():.2f} ± {subset['Potencia_Medida'].std():.2f}\n"
                  f"Estimada μ: {subset['Potencia_Estimada'].mean():.2f} ± {subset['Potencia_Estimada'].std():.2f}\n"
                  f"Δ Promedio: {(subset['Potencia_Estimada'] - subset['Potencia_Medida']).mean():.2f}")
    
    plt.text(0.02, 0.95, stats_text,
             transform=plt.gca().transAxes,
             bbox=dict(facecolor='white', alpha=0.8, edgecolor='#ddd'),
             verticalalignment='top')

plt.tight_layout()
plt.suptitle('Distribución de Potencia por Compresor', 
             y=1.02, fontsize=15, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='Compresor', y='Temperatura', data=df, palette=paleta)
plt.title('Distribución de Temperatura por Compresor (Violinplot)')
plt.grid(True)
plt.show()
```

```python
cols = ['Presion', 'Temperatura', 'Frecuencia', 'Potencia_Medida', 'Potencia_Estimada']
corr = df[cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de correlación entre variables numéricas')
plt.show()
```

FIREWALL


```python
firewall=pd.read_csv("../Datos/Transformados/logs.csv")
firewall
```

```python
top_ports = firewall['Destination Port'].value_counts().head(10)
plt.figure(figsize=(8, 4))
sns.barplot(x=top_ports.index.astype(str), y=top_ports.values, palette="Blues_d")
plt.title('Top 10 Puertos de Destino más Usados')
plt.xlabel('Puerto de destino')
plt.ylabel('Cantidad de conexiones')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Action', data=firewall, palette='Set2')
plt.title('Conteo de Acciones del Firewall')
plt.ylabel('Número de conexiones')
plt.xlabel('Acción')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=firewall, x='Bytes Sent', y='Bytes Received', alpha=0.6)
plt.title("Bytes Enviados vs Bytes Recibidos")
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
cols = ['Bytes', 'Bytes Sent', 'Bytes Received', 'Packets', 'Elapsed Time (sec)', 'pkts_sent', 'pkts_received']
corr = firewall[cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de correlación entre variables numéricas')
plt.show()
```
